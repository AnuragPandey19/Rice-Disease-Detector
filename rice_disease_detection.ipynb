```python
"""
Transfer Learning Implementation for Rice Leaf Disease Detection
Leveraging ResNet18 architecture pre-trained on ImageNet
"""

import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from tqdm import tqdm
import copy

# Import PyTorch modules
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import torchvision.models as models

# Initialize random seeds to ensure reproducible results
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)

#==============================================================================
# PROJECT SETTINGS
#==============================================================================

# Path and class definitions
BASE_DIR = "."  # Working directory
DISEASE_TYPES = ['Bacterial leaf blight', 'Brown spot', 'Leaf smut']
INPUT_DIMENSIONS = (224, 224)  # Standard ResNet input
TRAINING_BATCH = 16  # Batch size for training
TRAINING_CYCLES = 100  # Number of epochs
LR_VALUE = 0.001

# Dataset splitting proportions
TRAINING_SPLIT = 0.70
VALIDATION_SPLIT = 0.15
TESTING_SPLIT = 0.15

# Hardware setup
processing_unit = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Processing on: {processing_unit}\n")

#==============================================================================
# SECTION 1: ANALYZING THE DATASET
#==============================================================================

def analyze_dataset_structure(base_path, disease_categories):
    """Analyze and display dataset composition"""
    print("=" * 70)
    print("SECTION 1: DATASET STRUCTURE ANALYSIS")
    print("=" * 70)
    
    image_count = 0
    distribution_map = {}
    
    for category in disease_categories:
        category_folder = os.path.join(base_path, category)
        if os.path.exists(category_folder):
            image_files = [file for file in os.listdir(category_folder) 
                          if file.lower().endswith(('.jpg', '.jpeg', '.png'))]
            file_count = len(image_files)
            distribution_map[category] = file_count
            image_count += file_count
            print(f"  {category}: {file_count} images")
        else:
            print(f"  Alert: {category_folder} directory not found!")
    
    print(f"\n  Total Image Count: {image_count}")
    
    # Display warning for limited datasets
    if image_count < 100:
        print("\n  âš ï¸  NOTE: Dataset size is limited!")
        print("  â†’ Implementing Transfer Learning with ResNet18")
        print("  â†’ Extensive data augmentation techniques applied")
    
    print("=" * 70 + "\n")
    
    # Create distribution visualization
    plt.figure(figsize=(10, 6))
    category_names = list(distribution_map.keys())
    category_counts = list(distribution_map.values())
    
    color_scheme = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    plt.bar(category_names, category_counts, color=color_scheme)
    plt.xlabel('Disease Category', fontsize=12, fontweight='bold')
    plt.ylabel('Image Count', fontsize=12, fontweight='bold')
    plt.title('Distribution Across Classes', fontsize=14, fontweight='bold')
    plt.xticks(rotation=15, ha='right')
    
    for idx, count in enumerate(category_counts):
        plt.text(idx, count + 2, str(count), ha='center', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig('1_class_distribution.png', dpi=100, bbox_inches='tight')
    print("âœ“ Generated: 1_class_distribution.png\n")
    plt.show()
    
    return distribution_map

def display_sample_images(base_path, disease_categories, samples_count=4):
    """Show representative images from each category"""
    print("Displaying sample images from dataset...\n")
    
    fig, axes = plt.subplots(len(disease_categories), samples_count, 
                             figsize=(15, 4 * len(disease_categories)))
    
    for idx, category in enumerate(disease_categories):
        category_folder = os.path.join(base_path, category)
        
        if os.path.exists(category_folder):
            available_images = [file for file in os.listdir(category_folder) 
                               if file.lower().endswith(('.jpg', '.jpeg', '.png'))]
            chosen_images = random.sample(available_images, min(samples_count, len(available_images)))
            
            for col, image_file in enumerate(chosen_images):
                image_location = os.path.join(category_folder, image_file)
                loaded_image = Image.open(image_location)
                
                current_axis = axes[idx, col] if len(disease_categories) > 1 else axes[col]
                current_axis.imshow(loaded_image)
                current_axis.axis('off')
                
                if col == 0:
                    current_axis.set_title(f"{category}", fontsize=12, fontweight='bold')
    
    plt.tight_layout()
    plt.savefig('2_sample_images.png', dpi=100, bbox_inches='tight')
    print("âœ“ Generated: 2_sample_images.png\n")
    plt.show()

#==============================================================================
# SECTION 2: DATASET HANDLER
#==============================================================================

class LeafDiseaseDataset(Dataset):
    """Custom data handler for rice leaf images"""
    
    def __init__(self, paths_list, labels_list, transformation=None):
        self.paths_list = paths_list
        self.labels_list = labels_list
        self.transformation = transformation
    
    def __len__(self):
        return len(self.paths_list)
    
    def __getitem__(self, index):
        image_path = self.paths_list[index]
        loaded_image = Image.open(image_path).convert('RGB')
        image_label = self.labels_list[index]
        
        if self.transformation:
            loaded_image = self.transformation(loaded_image)
        
        return loaded_image, image_label

def gather_image_paths(base_path, disease_categories):
    """Collect all image paths with corresponding labels"""
    print("=" * 70)
    print("SECTION 2: GATHERING IMAGE PATHS")
    print("=" * 70)
    
    collected_paths = []
    collected_labels = []
    
    for label_id, category in enumerate(disease_categories):
        category_folder = os.path.join(base_path, category)
        print(f"  Processing {category}...")
        
        if os.path.exists(category_folder):
            found_images = [file for file in os.listdir(category_folder) 
                           if file.lower().endswith(('.jpg', '.jpeg', '.png'))]
            
            for image_file in found_images:
                full_path = os.path.join(category_folder, image_file)
                collected_paths.append(full_path)
                collected_labels.append(label_id)
    
    print(f"\n  Images collected: {len(collected_paths)}")
    print("=" * 70 + "\n")
    
    return collected_paths, collected_labels

#==============================================================================
# SECTION 3: IMAGE PREPROCESSING & AUGMENTATION
#==============================================================================

def prepare_transformations(dimensions, apply_augmentation=False):
    """Prepare image transformation pipeline"""
    if apply_augmentation:
        transformation_pipeline = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.RandomCrop(dimensions),
            transforms.RandomRotation(30),
            transforms.RandomHorizontalFlip(),
            transforms.RandomVerticalFlip(),
            transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.8, 1.2)),
            transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225])
        ])
    else:
        transformation_pipeline = transforms.Compose([
            transforms.Resize(dimensions),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225])
        ])
    
    return transformation_pipeline

def setup_data_loaders(paths_list, labels_list, train_portion, val_portion, test_portion, 
                       image_dims, batch_count):
    """Prepare data loaders for training, validation, and testing"""
    print("=" * 70)
    print("SECTION 3: PREPARING DATA LOADERS")
    print("=" * 70)
    
    # Initial split for training
    train_paths, temp_paths, train_labels, temp_labels = train_test_split(
        paths_list, labels_list, test_size=(1 - train_portion), 
        random_state=42, stratify=labels_list
    )
    
    # Split remaining data into validation and test
    adjusted_val = val_portion / (val_portion + test_portion)
    val_paths, test_paths, val_labels, test_labels = train_test_split(
        temp_paths, temp_labels, test_size=(1 - adjusted_val), 
        random_state=42, stratify=temp_labels
    )
    
    print(f"  Training samples: {len(train_paths)} ({train_portion*100:.0f}%)")
    print(f"  Validation samples: {len(val_paths)} ({val_portion*100:.0f}%)")
    print(f"  Testing samples: {len(test_paths)} ({test_portion*100:.0f}%)")
    
    # Apply transformations
    augmented_transform = prepare_transformations(image_dims, apply_augmentation=True)
    standard_transform = prepare_transformations(image_dims, apply_augmentation=False)
    
    training_data = LeafDiseaseDataset(train_paths, train_labels, transformation=augmented_transform)
    validation_data = LeafDiseaseDataset(val_paths, val_labels, transformation=standard_transform)
    testing_data = LeafDiseaseDataset(test_paths, test_labels, transformation=standard_transform)
    
    # Initialize loaders
    training_loader = DataLoader(training_data, batch_size=batch_count, 
                                shuffle=True, num_workers=0)
    validation_loader = DataLoader(validation_data, batch_size=batch_count, 
                                  shuffle=False, num_workers=0)
    testing_loader = DataLoader(testing_data, batch_size=batch_count, 
                               shuffle=False, num_workers=0)
    
    print("\n  Augmentation techniques applied:")
    print("    - Random cropping and resizing")
    print("    - Random rotation (up to 30 degrees)")
    print("    - Horizontal and vertical flipping")
    print("    - Color adjustments (brightness, contrast, saturation)")
    print("    - Affine transformations")
    print("    - Perspective variations")
    print("=" * 70 + "\n")
    
    return training_loader, validation_loader, testing_loader

#==============================================================================
# SECTION 4: MODEL ARCHITECTURE
#==============================================================================

def initialize_model(category_count):
    """Initialize transfer learning model using ResNet18"""
    print("=" * 70)
    print("SECTION 4: MODEL INITIALIZATION (ResNet18)")
    print("=" * 70)
    
    # Load pre-trained ResNet
    network = models.resnet18(pretrained=True)
    
    # Freeze initial layers
    for parameter in network.parameters():
        parameter.requires_grad = False
    
    # Modify final classification layer
    feature_count = network.fc.in_features
    network.fc = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(feature_count, 256),
        nn.ReLU(),
        nn.Dropout(0.3),
        nn.Linear(256, category_count)
    )
    
    network = network.to(processing_unit)
    
    # Calculate parameter counts
    all_params = sum(p.numel() for p in network.parameters())
    trainable = sum(p.numel() for p in network.parameters() if p.requires_grad)
    
    print(f"\n  Architecture: ResNet18 (Pre-trained on ImageNet)")
    print(f"  Total parameters: {all_params:,}")
    print(f"  Trainable parameters: {trainable:,}")
    print(f"  Frozen parameters: {all_params - trainable:,}")
    print(f"  Computing device: {processing_unit}")
    print("\n  Training approach:")
    print("  Phase 1: Train final layers only (20 epochs)")
    print("  Phase 2: Fine-tune complete network (remaining epochs)")
    print("=" * 70 + "\n")
    
    return network

#==============================================================================
# SECTION 5: TRAINING PROCESS
#==============================================================================

def execute_training(network, train_data, val_data, total_epochs, learn_rate):
    """Execute two-phase training process"""
    print("=" * 70)
    print("SECTION 5: TRAINING EXECUTION (Two-Phase Strategy)")
    print("=" * 70)
    
    loss_function = nn.CrossEntropyLoss()
    
    metrics_history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }
    
    highest_val_accuracy = 0.0
    optimal_weights = copy.deepcopy(network.state_dict())
    patience_count = 0
    max_patience = 15
    
    # PHASE 1: Train classifier only
    print("\n" + "="*70)
    print("PHASE 1: Training Classification Layer (First 20 epochs)")
    print("="*70)
    
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, network.parameters()), 
                          lr=learn_rate)
    lr_scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', 
                                                        factor=0.5, patience=5)
    
    initial_epochs = min(20, total_epochs // 2)
    
    for cycle in range(initial_epochs):
        # Training phase
        network.train()
        training_loss = 0.0
        correct_train = 0
        total_train = 0
        
        progress_bar = tqdm(train_data, desc=f'Phase1 Epoch {cycle+1}/{initial_epochs} [Train]')
        for batch_images, batch_labels in progress_bar:
            batch_images, batch_labels = batch_images.to(processing_unit), batch_labels.to(processing_unit)
            
            optimizer.zero_grad()
            predictions = network(batch_images)
            loss = loss_function(predictions, batch_labels)
            loss.backward()
            optimizer.step()
            
            training_loss += loss.item()
            _, predicted_classes = torch.max(predictions.data, 1)
            total_train += batch_labels.size(0)
            correct_train += (predicted_classes == batch_labels).sum().item()
            
            progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        training_loss /= len(train_data)
        training_accuracy = 100 * correct_train / total_train
        
        # Validation phase
        network.eval()
        validation_loss = 0.0
        correct_val = 0
        total_val = 0
        
        with torch.no_grad():
            for batch_images, batch_labels in val_data:
                batch_images, batch_labels = batch_images.to(processing_unit), batch_labels.to(processing_unit)
                predictions = network(batch_images)
                loss = loss_function(predictions, batch_labels)
                
                validation_loss += loss.item()
                _, predicted_classes = torch.max(predictions.data, 1)
                total_val += batch_labels.size(0)
                correct_val += (predicted_classes == batch_labels).sum().item()
        
        validation_loss /= len(val_data)
        validation_accuracy = 100 * correct_val / total_val
        
        lr_scheduler.step(validation_accuracy)
        
        metrics_history['train_loss'].append(training_loss)
        metrics_history['train_acc'].append(training_accuracy)
        metrics_history['val_loss'].append(validation_loss)
        metrics_history['val_acc'].append(validation_accuracy)
        
        print(f'Epoch {cycle+1}: Train Loss: {training_loss:.4f}, Train Acc: {training_accuracy:.2f}% | Val Loss: {validation_loss:.4f}, Val Acc: {validation_accuracy:.2f}%')
        
        if validation_accuracy > highest_val_accuracy:
            highest_val_accuracy = validation_accuracy
            optimal_weights = copy.deepcopy(network.state_dict())
            print(f'  âœ“ Optimal model saved (Val Acc: {validation_accuracy:.2f}%)')
            patience_count = 0
        else:
            patience_count += 1
    
    # PHASE 2: Fine-tune complete model
    print("\n" + "="*70)
    print("PHASE 2: Fine-tuning Complete Network")
    print("="*70)
    
    # Unfreeze all parameters
    for parameter in network.parameters():
        parameter.requires_grad = True
    
    # Lower learning rate for fine-tuning
    optimizer = optim.Adam(network.parameters(), lr=learn_rate / 10)
    lr_scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', 
                                                        factor=0.5, patience=5)
    
    remaining_cycles = total_epochs - initial_epochs
    
    for cycle in range(remaining_cycles):
        # Training phase
        network.train()
        training_loss = 0.0
        correct_train = 0
        total_train = 0
        
        progress_bar = tqdm(train_data, desc=f'Phase2 Epoch {cycle+1}/{remaining_cycles} [Train]')
        for batch_images, batch_labels in progress_bar:
            batch_images, batch_labels = batch_images.to(processing_unit), batch_labels.to(processing_unit)
            
            optimizer.zero_grad()
            predictions = network(batch_images)
            loss = loss_function(predictions, batch_labels)
            loss.backward()
            optimizer.step()
            
            training_loss += loss.item()
            _, predicted_classes = torch.max(predictions.data, 1)
            total_train += batch_labels.size(0)
            correct_train += (predicted_classes == batch_labels).sum().item()
            
            progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        training_loss /= len(train_data)
        training_accuracy = 100 * correct_train / total_train
        
        # Validation phase
        network.eval()
        validation_loss = 0.0
        correct_val = 0
        total_val = 0
        
        with torch.no_grad():
            for batch_images, batch_labels in val_data:
                batch_images, batch_labels = batch_images.to(processing_unit), batch_labels.to(processing_unit)
                predictions = network(batch_images)
                loss = loss_function(predictions, batch_labels)
                
                validation_loss += loss.item()
                _, predicted_classes = torch.max(predictions.data, 1)
                total_val += batch_labels.size(0)
                correct_val += (predicted_classes == batch_labels).sum().item()
        
        validation_loss /= len(val_data)
        validation_accuracy = 100 * correct_val / total_val
        
        lr_scheduler.step(validation_accuracy)
        
        metrics_history['train_loss'].append(training_loss)
        metrics_history['train_acc'].append(training_accuracy)
        metrics_history['val_loss'].append(validation_loss)
        metrics_history['val_acc'].append(validation_accuracy)
        
        print(f'Epoch {initial_epochs + cycle+1}: Train Loss: {training_loss:.4f}, Train Acc: {training_accuracy:.2f}% | Val Loss: {validation_loss:.4f}, Val Acc: {validation_accuracy:.2f}%')
        
        if validation_accuracy > highest_val_accuracy:
            highest_val_accuracy = validation_accuracy
            optimal_weights = copy.deepcopy(network.state_dict())
            torch.save(network.state_dict(), 'best_model.pth')
            print(f'  âœ“ Optimal model saved (Val Acc: {validation_accuracy:.2f}%)')
            patience_count = 0
        else:
            patience_count += 1
        
        if patience_count >= max_patience:
            print(f'\n  Early stopping activated after {initial_epochs + cycle+1} epochs')
            break
    
    # Load optimal model weights
    network.load_state_dict(optimal_weights)
    
    print("\n" + "=" * 70)
    print(f"âœ“ Training completed successfully!")
    print(f"  Best validation accuracy achieved: {highest_val_accuracy:.2f}%")
    print("  Model weights saved to 'best_model.pth'")
    print("=" * 70 + "\n")
    
    return metrics_history

#==============================================================================
# SECTION 6: PERFORMANCE EVALUATION
#==============================================================================

def visualize_training_metrics(metrics_history):
    """Generate training performance visualization"""
    print("Creating training history visualization...\n")
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    epoch_range = range(1, len(metrics_history['train_loss']) + 1)
    
    # Accuracy plot
    ax1.plot(epoch_range, metrics_history['train_acc'], label='Train Accuracy', linewidth=2)
    ax1.plot(epoch_range, metrics_history['val_acc'], label='Val Accuracy', linewidth=2)
    ax1.axvline(x=20, color='red', linestyle='--', label='Fine-tuning begins', alpha=0.7)
    ax1.set_title('Accuracy Progress', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Epoch', fontsize=12)
    ax1.set_ylabel('Accuracy (%)', fontsize=12)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Loss plot
    ax2.plot(epoch_range, metrics_history['train_loss'], label='Train Loss', linewidth=2)
    ax2.plot(epoch_range, metrics_history['val_loss'], label='Val Loss', linewidth=2)
    ax2.axvline(x=20, color='red', linestyle='--', label='Fine-tuning begins', alpha=0.7)
    ax2.set_title('Loss Progress', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Epoch', fontsize=12)
    ax2.set_ylabel('Loss', fontsize=12)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('3_training_history.png', dpi=100, bbox_inches='tight')
    print("âœ“ Generated: 3_training_history.png\n")
    plt.show()

def assess_model_performance(network, test_data, disease_categories):
    """Evaluate model on test dataset"""
    print("=" * 70)
    print("SECTION 6: TEST SET EVALUATION")
    print("=" * 70)
    
    network.eval()
    predicted_values = []
    actual_values = []
    probability_scores = []
    
    with torch.no_grad():
        for batch_images, batch_labels in tqdm(test_data, desc='Evaluating'):
            batch_images, batch_labels = batch_images.to(processing_unit), batch_labels.to(processing_unit)
            predictions = network(batch_images)
            probabilities = torch.nn.functional.softmax(predictions, dim=1)
            _, predicted_classes = torch.max(predictions, 1)
            
            predicted_values.extend(predicted_classes.cpu().numpy())
            actual_values.extend(batch_labels.cpu().numpy())
            probability_scores.extend(probabilities.cpu().numpy())
    
    predicted_values = np.array(predicted_values)
    actual_values = np.array(actual_values)
    probability_scores = np.array(probability_scores)
    
    # Calculate overall accuracy
    overall_accuracy = 100 * np.sum(predicted_values == actual_values) / len(actual_values)
    print(f"\n  Test Set Accuracy: {overall_accuracy:.2f}%\n")
    
    # Display detailed classification metrics
    print("  Detailed Classification Report:")
    print(classification_report(actual_values, predicted_values, target_names=disease_categories))
    
    # Generate confusion matrix
    conf_matrix = confusion_matrix(actual_values, predicted_values)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
                xticklabels=disease_categories, yticklabels=disease_categories, 
                cbar_kws={'label': 'Count'})
    plt.title('Confusion Matrix Analysis', fontsize=14, fontweight='bold')
    plt.ylabel('Actual Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()
    plt.savefig('4_confusion_matrix.png', dpi=100, bbox_inches='tight')
    print("\nâœ“ Generated: 4_confusion_matrix.png")
    plt.show()
    
    print("=" * 70 + "\n")
    
    return predicted_values, probability_scores, actual_values

def show_prediction_results(test_data, network, disease_categories, display_count=12):
    """Display model predictions on test images"""
    print("Creating prediction visualization...\n")
    
    network.eval()
    
    # Collect predictions
    image_batch = []
    label_batch = []
    prediction_batch = []
    probability_batch = []
    
    with torch.no_grad():
        for batch_images, batch_labels in test_data:
            images_gpu = batch_images.to(processing_unit)
            predictions = network(images_gpu)
            probabilities = torch.nn.functional.softmax(predictions, dim=1)
            _, predicted_classes = torch.max(predictions, 1)
            
            image_batch.extend(batch_images.cpu().numpy())
            label_batch.extend(batch_labels.cpu().numpy())
            prediction_batch.extend(predicted_classes.cpu().numpy())
            probability_batch.extend(probabilities.cpu().numpy())
            
            if len(image_batch) >= display_count:
                break
    
    # Prepare visualization
    samples_to_show = min(display_count, len(image_batch))
    sample_indices = list(range(samples_to_show))
    
    grid_rows = (samples_to_show + 3) // 4
    fig, axes = plt.subplots(grid_rows, 4, figsize=(16, 4*grid_rows))
    axes = axes.ravel() if grid_rows > 1 else [axes] if samples_to_show == 1 else axes
    
    for position, idx in enumerate(sample_indices):
        # Reverse normalization
        display_image = image_batch[idx].transpose(1, 2, 0)
        norm_mean = np.array([0.485, 0.456, 0.406])
        norm_std = np.array([0.229, 0.224, 0.225])
        display_image = norm_std * display_image + norm_mean
        display_image = np.clip(display_image, 0, 1)
        
        actual_class = disease_categories[label_batch[idx]]
        predicted_class = disease_categories[prediction_batch[idx]]
        confidence_score = probability_batch[idx][prediction_batch[idx]] * 100
        
        axes[position].imshow(display_image)
        axes[position].axis('off')
        
        title_color = 'green' if label_batch[idx] == prediction_batch[idx] else 'red'
        title_text = f"True: {actual_class}\nPred: {predicted_class}\nConf: {confidence_score:.1f}%"
        axes[position].set_title(title_text, fontsize=10, color=title_color, fontweight='bold')
    
    # Hide unused subplots
    for position in range(samples_to_show, len(axes)):
        axes[position].axis('off')
    
    plt.tight_layout()
    plt.savefig('5_predictions.png', dpi=100, bbox_inches='tight')
    print("âœ“ Generated: 5_predictions.png\n")
    plt.show()

#==============================================================================
# MAIN PROGRAM EXECUTION
#==============================================================================

def run_complete_pipeline():
    """Execute complete training and evaluation pipeline"""
    print("\n" + "=" * 70)
    print("ðŸŒ¾ RICE LEAF DISEASE CLASSIFICATION - RESNET18 TRANSFER LEARNING ðŸŒ¾")
    print("=" * 70 + "\n")
    
    # Step 1: Analyze dataset
    distribution = analyze_dataset_structure(BASE_DIR, DISEASE_TYPES)
    display_sample_images(BASE_DIR, DISEASE_TYPES)
    
    # Step 2: Gather image paths
    all_paths, all_labels = gather_image_paths(BASE_DIR, DISEASE_TYPES)
    
    # Step 3: Setup data loaders
    train_loader, val_loader, test_loader = setup_data_loaders(
        all_paths, all_labels, TRAINING_SPLIT, VALIDATION_SPLIT, TESTING_SPLIT,
        INPUT_DIMENSIONS, TRAINING_BATCH
    )
    
    # Step 4: Initialize model
    classification_model = initialize_model(len(DISEASE_TYPES))
    
    # Step 5: Train model
    training_metrics = execute_training(classification_model, train_loader, val_loader, 
                                       TRAINING_CYCLES, LR_VALUE)
    
    # Step 6: Visualize training progress
    visualize_training_metrics(training_metrics)
    
    # Load optimal model
    classification_model.load_state_dict(torch.load('best_model.pth'))
    
    # Step 7: Evaluate on test data
    predictions, probabilities, true_labels = assess_model_performance(
        classification_model, test_loader, DISEASE_TYPES
    )
    
    # Step 8: Visualize predictions
    show_prediction_results(test_loader, classification_model, DISEASE_TYPES)
    
    print("=" * 70)
    print("âœ… PIPELINE EXECUTION COMPLETED SUCCESSFULLY")
    print("=" * 70)
    print("\nOutput Files Generated:")
    print("  1. 1_class_distribution.png")
    print("  2. 2_sample_images.png")
    print("  3. 3_training_history.png")
    print("  4. 4_confusion_matrix.png")
    print("  5. 5_predictions.png")
    print("  6. best_model.pth (Trained ResNet18 weights)")
    print("=" * 70 + "\n")

if __name__ == "__main__":
    run_complete_pipeline()
```
